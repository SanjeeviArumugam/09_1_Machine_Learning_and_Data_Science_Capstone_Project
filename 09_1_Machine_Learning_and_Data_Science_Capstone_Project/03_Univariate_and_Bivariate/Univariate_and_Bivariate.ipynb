```python
import pandas as pd
import seaborn as sns
```

```python
dataset=pd.read_csv("Preprocessed_Car_Resale_Price_Data.csv")
```

```python
dataset.info()
```

```python
dataset.columns
```

```python
quan=[]
qual=[]
for columnName in dataset.columns:
    #print(columnName)
    if(dataset[columnName].dtype=='O'):
        #print("qual")
        qual.append(columnName)
    else:
        #print("quan")
        quan.append(columnName)
```

```python
quan
```

```python
qual # there is no qualititative data. because we have already preprocessd the input data
```

```python
dataset["engine_capacity"].mean()
```

```python
dataset.describe()
```

```python
descriptive=pd.DataFrame(index=["Mean","Median","Mode"],columns=quan)
for columnName in quan:
    descriptive[columnName]["Mean"]=dataset[columnName].mean()
    descriptive[columnName]["Median"]=dataset[columnName].median()
    descriptive[columnName]["Mode"]=dataset[columnName].mode()[0]
```

```python
import pandas as pd
```

```python
# Reading the Dataset
dataset=pd.read_csv("Preprocessed_Car_Resale_Price_Data.csv")
```

```python
dataset
```

```python
independent=dataset[['Car Name', 'Model Year', 'engine_capacity', 'Kms_driven', 'transmission_type', 'fuel_type', 'owner_type', 'body_type']]
dependent=dataset[['resale_price']]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
X_train
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.svm import SVR
regressor=SVR(kernel="rbf",C=1000)
regressor.fit(X_train,Y_train)
```

```python
regressor.intercept_
```

```python
regressor.n_support_
```

```python
regressor.support_
```

```python
y_pred=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(Y_test,y_pred)
```

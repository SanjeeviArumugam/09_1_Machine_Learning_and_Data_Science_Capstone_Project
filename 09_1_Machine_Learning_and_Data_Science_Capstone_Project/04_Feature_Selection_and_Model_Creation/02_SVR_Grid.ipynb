```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Reading the Dataset
dataset=pd.read_csv("Preprocessed_Car_Resale_Price_Data.csv")
```

```python
dataset
```

```python
independent=dataset[['Car Name', 'Model Year', 'engine_capacity', 'Kms_driven', 'transmission_type', 'fuel_type', 'owner_type', 'body_type']]
dependent=dataset[['resale_price']]
```

```python
#split into training set and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(independent, dependent, test_size = 1/3, random_state = 0)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
param_grid = {'kernel':['rbf','sigmoid','linear'],
'C':[10,100,1000,2000,3000],'gamma':['auto','scale']}
grid = GridSearchCV(SVR(), param_grid, refit = True, verbose= 3,n_jobs=-1)
grid.fit(independent, dependent)
```

```python
re=grid.cv_results_
grid_predictions = grid.predict(independent)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(dependent,grid_predictions)
print("The R_score value for best parameter {}:".format(grid.best_params_),r_score)
```
